Import libraries that we need to do EDA

```python
import pandas as pd 
import numpy as np
```

```python
# read data netflix.csv
df = pd.read_csv("netflix.csv")
```

```python
# check info data
df.info()
```

- From that info we know, There are 8790 entries data and each column has 8790 entries that means there are no missing value
- In dtype we know about data type of each column. Then we check, is the data type true or not
- After analysis the data type in each column, The data type of data_added is object we should change to date and in column duration we should change the data type to int but before that we should split to 2 categorical : season and minute

```python
df['date_added']=pd.to_datetime(df['date_added']) # change data type in column data_added to date type
```

```python
# # split column duration to 2 categorical : season and minute

# # Making column in minute category
df['duration_minute'] = ""
for i,categori in enumerate(df['duration']):
  if 'min' in categori:
        df.at[i, 'duration_minute'] = categori
  elif 'season' in categori:
        df.at[i, 'duration_minute'] = pd.NA

# Change duration_minute to integer data type
# Make sure you have replaced string 'min' in that column, then you can convert to int
df['duration_minute'] = df['duration_minute'].str.replace(' min','')
df['duration_minute'] = df['duration_minute'].replace('', pd.NA)  # replace empty string with Nan
df['duration_minute'] = df['duration_minute'].astype(pd.Int32Dtype())
```

```python
df
```

```python
# Making column in season category
#df['duration_season'] = ""
df['duration_season'] = pd.NA

for i,categori in enumerate(df['duration']):
  if 'Season' in categori:
        #df.at[i, 'duration_season'] = categori
        df.at[i, 'duration_season'] = int(categori.replace(' Seasons', '').replace(' Season', ''))
  elif 'min' in categori:
        df.at[i, 'duration_season'] = pd.NA

# Change duration_season to integer data type
df['duration_season'] = df['duration_season'].astype(pd.Int32Dtype())
```

```python
df
```

```python
df.info()
```

- After we split duration column to be 2 parts : 'duration_minute' and 'duration_season', then we drop duration column
- In duration_minute and duration_column there are missing value cause the entries are not full 8790 but it's ok cause it's shared to be 2 columns

```python
# drop column 'duration'
df.drop(["duration"],axis=1, inplace = True)
```
